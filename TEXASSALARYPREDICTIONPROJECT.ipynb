_**REMOVING THE OUTLIERS**_

- . _**IQR**_
- . _IQR stand for the interquatile range, When a data set has outliers or extreme values, we summarize a typical value using
- . the median as opposed to the mean. which is the difference between the first and third quartiles_

```python
#for outlier handling Importing stats
from scipy import stats
```

_MONTHLY_

```python
iqr = stats.iqr(data['MONTHLY'],interpolation='midpoint')
print("IQR:",iqr)

# Get first and third quantile
Q1 = data['MONTHLY'].quantile(0.25)
Q3 = data['MONTHLY'].quantile(0.75)

# Get manimum and maximum limit
min_limit = Q1 - 1.5*iqr
print("Minimum limit:",min_limit)

max_limit = Q3 +1.5*iqr
print("Maximum limit",max_limit)
```

```python
data.loc[data['MONTHLY']<min_limit] #checking values which are less than minimum limit
```

```python
data.loc[data['MONTHLY']>max_limit]#checking values which are greater than maximum limit
```

```python
data.loc[data['MONTHLY']>max_limit,'MONTHLY']=np.median(data.MONTHLY)#imputing value with median
data.loc[data['MONTHLY']>max_limit]#checking whether outlier remove or not
```
